```python
#1 Tokenization
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt_tab')
```

```python
from nltk.tokenize import  word_tokenize
text = "Natural language processing is interesting"
tokens = word_tokenize(text)
print(tokens)
```

```python
# sentence tokenization
from nltk.tokenize import  sent_tokenize

text = "python is easy it is widely used in data science"
sentences = sent_tokenize(text)
print(sentences)
```

```python
# stop world removal
# stop words such as is,the,and,in are removed
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
tokens = word_tokenize(text)
filtered_tokens = [word for word in tokens if word.lower() not in stop_words]
print(filtered_tokens)

```

```python
# stemming
#stemming reduces word to their root form

from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
words = ["playing","study","running","gone","watched","coding"]
for w in words:
  print(stemmer.stem(w))
```

```python
#Lemmatization
#Lemmatization converts words to their dictionary form

from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

print(lemmatizer.lemmatize("writing"))
print(lemmatizer.lemmatize("running"))
print(lemmatizer.lemmatize("gone"))
print(lemmatizer.lemmatize("watched"))
print(lemmatizer.lemmatize("coding"))
```

```python
#Part-pf-speech (POS) Tagging
#POS tagging identifies grammatical roles of words

import nltk
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
from nltk import pos_tag
words = word_tokenize("Python is a programming language")
pos_tags = pos_tag(words)
print(pos_tags)
```

```python
# Word Frequency count
from nltk.probability import FreqDist

words = word_tokenize("Python is a programming language")
freq_dist = FreqDist(words)
print(freq_dist.most_common())
```
